# **Análisis Europa🐼📊🗺**

```python
import matplotlib.pyplot as plt

# Seleccionar solo algunas columnas específicas
years_columns = ['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2015 Population', '2020 Population', '2022 Population']

# Filtrar el DataFrame para obtener solo los datos de Europa
data_europa = df[df['Continent'] == 'Europe']

# Lista para almacenar las listas de población de cada país
poblacion_por_pais = []

# Iterar sobre cada país de Europa y obtener los datos de población
for pais in data_europa['Country/Territory'].unique():
    pais_df = data_europa[data_europa['Country/Territory'] == pais]
    poblacion_pais = pais_df[years_columns].values.tolist()[0]
    poblacion_por_pais.append(poblacion_pais)

# Lista de años
years = [1970, 1980, 1990, 2000, 2010, 2015, 2020, 2022]

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Iterar sobre los datos de población por país y graficarlos
for i, poblacion in enumerate(poblacion_por_pais):
    ax.plot(years, poblacion, label=data_europa['Country/Territory'].unique()[i])

# Mover la leyenda fuera de la gráfica
ax.legend(bbox_to_anchor=(1, 1), loc='upper left')

# Configurar etiquetas y título
ax.set_xlabel('Año')
ax.set_ylabel('Población')
ax.set_title('Población por país en Europa')

# Mostrar la gráfica
plt.show()
```
