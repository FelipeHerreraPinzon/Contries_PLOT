```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python


# Cargar los datos desde tu base de datos
df = pd.read_csv('https://raw.githubusercontent.com/FelipeHerreraPinzon/Data_Sharks_Project/main/world_population.csv')
df
```

```python
cat = ["bored", "happy", "none", "bored", "happy", "none"]
dog = ["none", "happy", "happy", "happy", "bored", "bored"]
bird = ["bored", "happy", "bored", "happy", "bored", "bored"]
hamster = ["happy", "happy", "bored", "happy", "bored", "bored"]

activity = ["combing", "drinking", "feeding", "napping", "playing", "washing"]

fig, ax = plt.subplots()
ax.plot(activity, dog, label="dog")
ax.plot(activity, bird, label="bird")
ax.plot(activity, cat, label="cat")
ax.plot(activity, hamster, label="hamster")
ax.legend()
# Mueve la leyenda fuera de la gráfica
ax.legend(bbox_to_anchor=(1, 1), loc='upper left')

plt.show()
```

```python
# DATA-FRAMES
colombia = df[df['Country/Territory'] == 'Colombia']
india = df[df['Country/Territory'] == 'India']
chile = df[df['Country/Territory'] == 'Chile']

# Seleccionar solo algunas columnas específicas
years_columns = ['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2015 Population', '2020 Population', '2022 Population']

# COLOMBIA
colombia_population = colombia[years_columns]
# Convertir los datos de población de Colombia en una lista
colombia_list = colombia_population.values.tolist()[0]

# INDIA
india_population = india[years_columns]
# Convertir los datos de población de Colombia en una lista
india_list = india_population.values.tolist()[0]

# CHILE
chile_population = chile[years_columns]
# Convertir los datos de población de Colombia en una lista
chile_list = chile_population.values.tolist()[0]


years = ["1970", "1980", "1990", "2000", "2010", "2015", "2020", "2022"]

fig, ax = plt.subplots()
ax.plot(years, colombia_list, label="Colombia")
#ax.plot(years, india_list, label="India")
ax.plot(years, chile_list, label="Chile")

ax.legend()
# Mueve la leyenda fuera de la gráfica
ax.legend(bbox_to_anchor=(1, 1), loc='upper left')

plt.show()
```
